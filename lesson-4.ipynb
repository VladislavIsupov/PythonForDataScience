Задание 2
С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.
Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.
Постройте линейную диаграмму, используя массив t для координат по горизонтали,
а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.
Выведите название диаграммы - 'График f(t)'.
Также добавьте названия для горизонтальной оси - 'Значения t'
и для вертикальной - 'Значения f'.
Ограничьте график по оси x значениями 0.5 и 9.5,
а по оси y - значениями -2.5 и 2.5.

```python
t = np.linspace(0, 10, num=51, endpoint=True)
t
```

```python
f = np.cos(t)
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "Gray",
    "family": "serif",
}

label_font = {
    "fontsize": 10,
    "family": "serif",
    "color": "#505050",
}

legend_font = {
    "size": 7,
    "family": "serif",
}
```

```python
plt.plot(t, f, color = 'green', label = 'cos(x)')

plt.axis([.5, 9.5, -2.5, 2.5])

plt.title('График f(t)', fontdict=title_font)
plt.xlabel('Значения t', fontdict=label_font)
plt.ylabel('Значения f', fontdict=label_font)

plt.grid(axis="both", color="lightgrey", linestyle = 'dashed')

legend = plt.legend(loc = 'upper left', frameon=False, prop=legend_font)
plt.setp(legend.get_texts(), color="Gray")

plt.show
```

Задание *3
С помощью функции linspace библиотеки Numpy создайте массив x
из 51 числа от -3 до 3 включительно.
Создайте массивы y1, y2, y3, y4 по следующим формулам:
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = sin(x)
Используя функцию subplots модуля matplotlib.pyplot,
создайте объект matplotlib.figure.Figure с названием fig 
и массив объектов Axes под названием ax,
причем так, чтобы у вас было 4 отдельных графика в сетке,
состоящей из двух строк и двух столбцов.
В каждом графике массив x используется для координат по горизонтали.
В левом верхнем графике для координат по вертикали используйте y1,
в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4.
Дайте название графикам: 'График y1', 'График y2' и т.д.
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
x = np.linspace(-3, 3, num = 51, endpoint = 'True')
```

```python
y1 = x**2
y2 = 2 * x + .5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_xlim([-5, 5])
ax1.set_title("График y1")

ax2.plot(x, y2)
ax2.set_title("График y1")

ax3.plot(x, y3)
ax3.set_title("График y1")

ax4.plot(x, y4)
ax4.set_title("График y1")
```

```python
import pandas as pd
```

```python
creditcard = pd.read_csv('creditcard.csv')
```

```python
plt.style.use('fivethirtyeight')
```

```python
class_counts = creditcard['Class'].value_counts()
class_counts
```

```python
class_counts.plot(kind='bar', logy=False)
plt.show()
```

```python
class_counts.plot(kind='bar', logy=True)
plt.show()
```

```python
plt.hist(creditcard.query('Class == 0')["V1"], bins=20, color = 'grey', ec='gray',\
         log=False, density = 'True', alpha = .5, label="Class 0")
plt.hist(creditcard.query('Class == 1')["V1"], bins=20, color = 'red', ec='gray', \
         log = False, density = 'True', alpha = .5, label="Class 1")
plt.legend()
plt.xlabel('Class')
plt.show()
```
